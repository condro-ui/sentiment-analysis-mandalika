```python
df_full = pd.concat([df1,df2,df3,df4,df5,df6,df7,df8,df9,df10], axis = 0)
```

```python
import pandas as pd

# Asumsikan df_full sudah di-load sebagai DataFrame pandas
# df_full = pd.read_csv('path_to_your_file.csv')

# --- Langkah 1: Identifikasi Kunci Unik ---
# Kita memilih 'id_str' sebagai kunci unik untuk setiap tweet.
unique_key = 'id_str'
print(f"🔍 Menggunakan kolom '{unique_key}' sebagai basis pengecekan duplikat.\n")

# --- Langkah 2: Deteksi Data Duplikat ---
# Method .duplicated() akan mengembalikan boolean Series (True jika duplikat, False jika tidak).
# Kita gunakan parameter 'subset' untuk memeriksa duplikasi hanya pada kolom kunci.
# .sum() akan menjumlahkan semua nilai True untuk mendapatkan total baris duplikat.

duplicate_rows = df_full.duplicated(subset=[unique_key]).sum()

if duplicate_rows > 0:
    print(f"🔴 Ditemukan {duplicate_rows} baris data duplikat.")
    print(f"Dimensi dataframe awal (sebelum transformasi): {df_full.shape}")

    # --- Langkah 3: Transformasi (Menghapus Duplikat) ---
    # Kita gunakan method .drop_duplicates() untuk membuat dataframe baru yang unik.
    # subset=[unique_key]: Menentukan basis kolom untuk penghapusan.
    # keep='first': Mempertahankan baris pertama yang muncul dan menghapus duplikat setelahnya.
    # inplace=False (default): Membuat dataframe baru, tidak memodifikasi dataframe asli.

    df_unique = df_full.drop_duplicates(subset=[unique_key], keep='first')

    print(f"Dimensi dataframe setelah transformasi: {df_unique.shape}")

    # --- Langkah 4: Verifikasi ---
    # Memastikan tidak ada lagi duplikat di dataframe baru.
    remaining_duplicates = df_unique.duplicated(subset=[unique_key]).sum()
    print(f"✅ Verifikasi: Jumlah duplikat setelah dibersihkan adalah {remaining_duplicates}.\n")

    # Anda sekarang dapat menggunakan df_unique untuk analisis selanjutnya
    # Contoh: print(df_unique.info())

else:
    print("✅ Selamat! Tidak ditemukan data duplikat berdasarkan kolom 'id_str'.")
    print(f"Dataframe Anda sudah unik dengan {df_full.shape[0]} baris data.")
    df_unique = df_full # Jika tidak ada duplikat, df_unique merujuk ke df_full
```
